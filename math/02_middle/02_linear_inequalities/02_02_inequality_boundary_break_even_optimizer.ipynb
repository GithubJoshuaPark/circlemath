# 🟢 2단계: 일차부등식 (Linear Inequalities)
> **완벽한 평등 너머의 포용적 영역, 그리고 음수(Negative)가 부르는 부등호의 반전**

---

### 1. 묵상과 사유 (철학적·종교적 관점)
방정식(Equation)이 좌변과 우변이 완벽히 같아야만 문이 열리는 단 하나의 한 점($=$)을 구하는 엄격함의 세계라면, 부등식(Inequality)은 부등호($<, >$)를 활용해 넉넉하게 열려 있는 가능성의 **'영역(Domain)'**을 획정하는 세계입니다.

* **음수(Negative)를 통과할 때 일어나는 가치의 반전**
  부등식의 성질 중 가장 핵심적이고 신비로운 것은 **"양변에 음수를 곱하거나 나눌 때 부등호의 방향이 바뀐다"**는 법칙입니다. 양수로 가득 찬 밝은 세상에서는 크고 강한 것들이 늘 앞서지만, 어둡고 부정적인 것(음수)을 마주하거나 통과할 때는 기존의 선후 관계와 판단의 질서가 정반대로 뒤집힙니다.
  철학적으로 이는 삶의 시련이나 부정적 위기(음수)를 만났을 때, 우리가 가졌던 기존의 가치 기준과 강함의 우선순위가 180도 역전되는 내면적 성찰을 떠올리게 합니다. "첫째가 꼴찌가 되고 꼴찌가 첫째가 된다"는 종교적 역설처럼, 약함과 고난(음수)이 오히려 진정한 방향(부등호)을 바로잡는 기회가 되는 섭리를 보여줍니다.

* **한 점의 정답이 아닌, 은혜의 넉넉한 영역**
  부등식의 해는 단 하나의 고정된 수치가 아닌 '수직선 상의 빗금 친 영역'으로 표현됩니다. 어떤 절대적이고 완벽한 한 점에 도달하지 못하더라도, 약속된 기준선(경계)을 넘어서 존재한다면 모두 용납되는 포용의 미학을 담고 있습니다. 
  인생 and 신앙의 여정 역시 단 하나의 오차도 없는 완벽한 정답(등식)에 도달하는 일보다, 신성한 기준의 영역(부등식) 안에 머무르며 누리는 넉넉한 은혜와 다양성을 누리는 데 있음을 묵상하게 합니다.

* **안전을 지키는 선율: 경계(Boundary)의 축복**
  부등식은 $x \le 50$과 같이 존재가 넘어서는 안 될 경계선과 한계값을 명확히 그어줍니다. 이 경계는 존재를 억압하기 위한 쇠창살이 아니라, 그 안에서 구성원이 가장 안전하고 자유롭게 숨 쉴 수 있도록 보호해 주는 안전막입니다. 경계의 존재는 한계에 순응하는 겸손함이 곧 참된 평화의 출발점임을 가르쳐 줍니다.

---

### 2. 왜 사용하는가? 실제 생활에서의 적용점

* **생명과 재난을 막아내는 공학적 약속: 허용 한계(Tolerance)와 안전율**
  승강기의 탑승 중량 제한($W \le 800\text{kg}$), 도로 터널의 높이 제한, 다리가 버틸 수 있는 최대 하중 설계 등 모든 안전 공학의 공식은 방정식이 아닌 부등식으로 설계됩니다. "정확히 $800\text{kg}$이어야 한다"가 아니라, "그 이하에 머물러야 안전하다"라는 부등식의 여유 공간(Buffer)이 예측 불가능한 재난으로부터 매일 수많은 사람의 생명을 구해내고 있습니다.

* **비즈니스의 최적화와 리스크 관리: 손익분기점(BEP) 돌파**
  기업 경영에서 가장 기본이 되는 수식은 `매출 > 비용` 이라는 부등식입니다. 매출이 고정비와 가변비의 합을 넘어서 손익분기점을 돌파하는 순간, 기업은 비로소 생존과 흑자라는 새로운 영역으로 진입하게 됩니다. 여러 리스크 요인들의 한계치 조건(부등식)을 결합하여 자원을 최적으로 분배하는 '선형 계획법(Linear Programming)'의 기초가 바로 부등식의 연산입니다.

* **디지털 논리 흐름을 제어하는 조건문: `if (condition)`**
  현대 소프트웨어 코딩에서 흐름을 분기하는 제어문(`if-else`)의 거의 대부분은 부등식 조건을 평가합니다. 센서 측정값이 특정 임계값보다 커질 때 알림을 울리거나, 계정 잔액이 인출 요청 금액보다 클 때 거래를 승인하는 등 정보 사회의 정교한 시스템 제어가 부등식을 통해 실행됩니다.

## 1. 일차부등식 $ax + b > c$ 경계 시각화 및 반전 체험 (Inequality Boundary)

아래 실습은 수직선 위에 일차부등식 $ax + b > c$의 해를 실시간 파란색 영역(Shaded Region)으로 표현하고, 계수 $a$의 슬라이더를 잡고 음수 영역으로 내릴 때, 영역이 어떻게 순식간에 반대 방향으로 반전(Flip)하는지 눈으로 체험하는 대화형 수직선 기하 툴입니다.
경계값 포함 여부에 따라 경계점이 꽉 찬 원과 빈 원으로 바뀌는 수학적 규칙도 함께 관찰해 보세요.

### 🛠️ 1단계: 도구 준비 (Imports & Settings)
실습을 진행하기 위해 필요한 파이썬 라이브러리(`numpy`, `matplotlib.pyplot`, `ipywidgets` 등)를 불러오고, 그래프에서 한글 폰트가 깨지지 않도록 한글 환경 설정을 수행합니다.

```python
import platform
import numpy as np  # 대규모 다차원 배열과 행렬 연산을 위한 라이브러리입니다.
import matplotlib.pyplot as plt  # 데이터를 시각화(그래프 그리기)하기 위한 핵심 도구입니다.
import ipywidgets as widgets  # 슬라이더, 버튼 등 대화형 UI 구성 요소를 만드는 라이브러리입니다.
from ipywidgets import interact  # 슬라이더 조작에 따라 특정 함수를 즉시 다시 실행시켜주는 인터랙티브 데코레이터 함수입니다.

# 주피터 노트북 및 matplotlib 시각화 환경에서 OS별 한글 폰트가 깨지는 현상을 방지하기 위한 초기 설정입니다.
if platform.system() == "Darwin":  # 사용자의 OS가 macOS(Darwin)일 경우
    plt.rcParams["font.family"] = "AppleGothic"  # 애플 운영체제 기본 한글 폰트인 AppleGothic을 적용합니다.
elif platform.system() == "Windows":  # 사용자의 OS가 Windows일 경우
    plt.rcParams["font.family"] = "Malgun Gothic"  # 마이크로소프트 기본 한글 폰트인 맑은 고딕을 적용합니다.

# 그래프의 축 좌표 범위 등에 마이너스(-) 기호가 사각형 등으로 깨지는 에러를 방지합니다.
plt.rcParams["axes.unicode_minus"] = False
```

### 📈 2단계: 수학적 데이터 생성 및 그리기 함수 정의 (Data Generation & Functions)
입력받은 계수 $a$, 상수 $b$, 기준값 $c$ 및 부등호 연산자를 기반으로 부등식 해의 경계점을 계산하고, 수직선 기하 위에 실시간 해의 영역을 드로잉하는 `plot_inequality` 함수를 정의합니다.

```python
def plot_inequality(a, b, c, op):
    # matplotlib figure를 생성합니다. 수직선만 그릴 것이므로 세로 길이를 작게(10 x 3.5) 잡습니다.
    fig, ax = plt.subplots(figsize=(10, 3.5))
    
    # x축의 범위를 -10부터 10까지로 고정하여 경계점의 움직임을 상대적으로 관찰하기 쉽게 만듭니다.
    ax.set_xlim(-10, 10)
    # y축은 수직선 주위의 여백을 주기 위해 -1부터 1까지로 제한합니다.
    ax.set_ylim(-1, 1)
    
    # 불필요한 테두리(상, 하, 좌, 우)를 숨겨 수직선만 돋보이게 디자인합니다.
    for spine in ["top", "bottom", "left", "right"]:
        ax.spines[spine].set_visible(False)
        
    # y축의 눈금(ticks)과 레이블을 제거하여 수직선 단 하나에만 집중할 수 있게 합니다.
    ax.set_yticks([])
    
    # x축 눈금과 레이블은 유지하되, y=0 선상에 수직선 역할을 할 회색 가로선을 굵게 그립니다.
    ax.axhline(0, color="gray", linewidth=2, zorder=1)
    
    # x축의 눈금 위치를 설정하여 정수 단위 구분을 명확히 합니다.
    ax.set_xticks(range(-10, 11))
    ax.grid(True, axis='x', linestyle='--', alpha=0.5)

    # a가 0인 특수한 경우의 예외 처리를 진행합니다. 0으로 나누는 수학적 에러를 차단하기 위함입니다.
    if a == 0:
        # a = 0일 때, 식은 0*x + b op c -> b op c 가 됩니다. 이 식의 참/거짓 여부를 평가합니다.
        op_map = {">": ">", ">=": ">=", "<": "<", "<=": "<="}
        is_true = eval(f"{b} {op_map[op]} {c}")
        
        if is_true:
            # 식이 항상 참일 경우, 수직선의 모든 영역을 파란색 해 영역으로 칠합니다.
            ax.axhspan(-0.3, 0.3, color="#0B79D0", alpha=0.2, label="해: 모든 실수 (Always True)")
            title_text = f"현재 식: {a}x + ({b}) {op} {c} $\\to$ {b} {op} {c} (참)\n결과: 모든 실수가 해가 됩니다."
        else:
            # 식이 항상 거짓일 경우, 수직선에 칠을 하지 않고 해가 없다는 안내를 띄웁니다.
            title_text = f"현재 식: {a}x + ({b}) {op} {c} $\\to$ {b} {op} {c} (거짓)\n결과: 해가 존재하지 않습니다."
            
        ax.set_title(title_text, fontsize=12, pad=20, color="#333333", weight="bold")
        ax.legend(loc="upper right")
        plt.tight_layout()
        plt.show()
        return

    # a가 0이 아닐 경우, 정상적으로 경계값을 연산합니다.
    # 경계값 boundary_val = (c - b) / a
    boundary_val = (c - b) / a
    
    # 연산식 방향(Flip)을 결정하는 핵심 수학 로직입니다.
    # a의 부호에 따라 부등호의 방향이 바뀔지 아닐지를 결정합니다.
    flip_map = {
        ">": "<",
        ">=": "<=",
        "<": ">",
        "<=": ">="
    }
    
    # 실제 부등호 방향을 결정합니다. a가 음수이면 뒤집고, 양수이면 그대로 사용합니다.
    actual_op = flip_map[op] if a < 0 else op
    is_inclusive = "=" in op  # 경계값 자체를 포함하는지 여부 (>=, <= 이면 True)
    
    # 칠할 영역의 범위를 구합니다.
    # actual_op가 '>' 나 '>=' 이면 경계값부터 무한대(여기서는 그래프 범위 바깥인 15)까지 칠합니다.
    # actual_op가 '<' 나 '<=' 이면 마이너스 무한대(여기서는 -15)부터 경계값까지 칠합니다.
    if ">" in actual_op:
        # 경계값 x_boundary의 오른쪽 영역을 칠합니다.
        ax.axvspan(boundary_val, 15, color="#0B79D0", alpha=0.2, label="부등식의 해 영역 (Solution)")
        # 방향을 가리키는 굵은 파란색 화살표를 렌더링합니다.
        ax.annotate("", xy=(9, 0), xytext=(boundary_val, 0),
                    arrowprops=dict(arrowstyle="->", color="#0B79D0", lw=3.5, mutation_scale=20))
    else:
        # axvspan을 사용해 경계값 x_boundary의 왼쪽 영역을 칠합니다.
        ax.axvspan(-15, boundary_val, color="#0B79D0", alpha=0.2, label="부등식의 해 영역 (Solution)")
        # 방향을 가리키는 굵은 파란색 화살표를 렌더링합니다.
        ax.annotate("", xy=(-9, 0), xytext=(boundary_val, 0),
                    arrowprops=dict(arrowstyle="->", color="#0B79D0", lw=3.5, mutation_scale=20))
                    
    # 경계점 자체의 시각적 형태를 지정합니다. (속이 빈 원 vs 채워진 원)
    # 초등/중등 교육과정 기하 표현 규칙에 완벽하게 일치시킵니다.
    if is_inclusive:
        # 경계값을 포함하므로 속이 꽉 찬 파란색 원으로 채웁니다.
        ax.plot(boundary_val, 0, marker="o", color="#0B79D0", markersize=12, zorder=5, label=f"경계 포함: x = {boundary_val:.2f}")
    else:
        # 경계값을 포함하지 않으므로 속이 비어 있는(배경색과 같은 흰색으로 채운) 원으로 표현합니다.
        ax.plot(boundary_val, 0, marker="o", color="#0B79D0", mfc="white", mec="#0B79D0", mew=2.5, markersize=12, zorder=5, label=f"경계 제외: x = {boundary_val:.2f}")

    # LaTeX를 사용해 실시간 부등식 풀이 전개 과정을 동적으로 렌더링합니다.
    # a의 음수/양수 여부에 따라 부등호 방향 전환에 대한 텍스트 설명을 추가하여 교육적 가치를 제공합니다.
    step1 = f"{a}x + ({b}) {op} {c}"
    step2 = f"{a}x {op} {c - b}"
    
    if a < 0:
        explain_txt = f"[경고] 양변을 음수 $a = {a}$로 나누므로 부등호 방향이 반전(Flip)됩니다!"
        step3 = f"x {actual_op} \\frac{{{c - b}}}{{{a}}} \\to x {actual_op} {boundary_val:.2f}"
    else:
        explain_txt = f"[안내] 양변을 양수 $a = {a}$로 나누므로 부등호 방향이 그대로 유지됩니다."
        step3 = f"x {actual_op} \\frac{{{c - b}}}{{{a}}} \\to x {actual_op} {boundary_val:.2f}"
        
    title_text = f"수식 전개: ${step1} \\to {step2} \\to {step3}$\n{explain_txt}"
    ax.set_title(title_text, fontsize=11, pad=20, color="#333333", weight="bold")
    
    # 범례(Legend)를 표시하고 레이아웃을 조절합니다.
    ax.legend(loc="upper right", frameon=True, facecolor="white", edgecolor="none")
    plt.tight_layout()
    plt.show()
```

### 🎛️ 3단계: 인터랙티브 슬라이더 실행 및 실시간 시각화 (Plotting & Execution)
앞서 정의한 그리기 함수를 호출하여 인터랙티브 대시보드를 생성합니다. 슬라이더 $a$, $b$, $c$를 마우스로 조작해 보면서, 특히 **$a$가 양수에서 음수 영역으로 넘어갈 때** 해의 영역 방향(화살표)과 수식 기호가 순간적으로 반대 방향으로 뒤집히는 물리적 질서의 반전을 느껴보세요.

```python
# interact 함수를 가동하여 plot_inequality 함수에 widgets.FloatSlider 및 Dropdown을 동적으로 연동합니다.
# a의 범위를 -10부터 10까지 0.5 단위로 변경할 수 있도록 슬라이더를 구성합니다.
interact(
    plot_inequality,
    a=widgets.FloatSlider(value=2.0, min=-10.0, max=10.0, step=0.5, description="계수 a (기울기)"),
    b=widgets.FloatSlider(value=4.0, min=-10.0, max=10.0, step=0.5, description="상수 b (절편)"),
    c=widgets.FloatSlider(value=8.0, min=-10.0, max=10.0, step=0.5, description="기준 c (목표값)"),
    op=widgets.Dropdown(options=[">", ">=", "<", "<="], value=">", description="부등호 op")
)
```

## 2. 컴퓨터 비즈니스 의사결정의 뼈대: 손익분기점(BEP) 최적화 대시보드 (Break-even Optimizer)

일차부등식은 단순한 교과서 수학 문제를 넘어, 기업 경영의 생존을 결정짓는 핵심 도구입니다.
가장 대표적인 사례가 바로 **'손익분기점(Break-even Point, BEP)'** 계산입니다. 기업이 살아남기 위해서는 `매출 > 비용` 이라는 부등식이 참이어야 합니다. 즉, 이익이 0보다 큰 영역($\text{Profit} > 0$)을 확보하기 위해 판매량이 최소한 어느 한계값(임계점)을 초과해야 하는지를 빗금과 실시간 추세선 그래프로 시뮬레이션해 봅니다.

* **수학적 모델링**
  - 판매량 (사용자 수) = $x$ 명
  - 고정비 (Fixed Cost) = $F$ 만 원 (인프라 서버 비용, 기본 유지비 등)
  - 판매 단가 (Price) = $P$ 만 원 (SaaS 월 구독료 등)
  - 단위당 변동비 (Variable Cost) = $V$ 만 원 (사용자 1명당 API 사용 수수료 등)
  - **부등식 조건**: $\text{Revenue} > \text{Cost} \\implies P \cdot x > F + V \cdot x \\implies (P - V)x > F$
  - **최소 돌파 임계값**: $x > \frac{F}{P - V}$

### 🛠️ 1단계: 도구 준비 (Imports & Settings)
이 실습 주제만 단독으로 다시 실행하더라도 라이브러리나 폰트 설정 누락으로 오류가 나지 않도록, 필요한 모듈 임포트와 한글 설정을 완전히 내장하여 독립성을 보장합니다.

```python
import platform
import numpy as np  # 대규모 다차원 배열과 행렬 연산을 위한 라이브러리입니다.
import matplotlib.pyplot as plt  # 데이터를 시각화(그래프 그리기)하기 위한 핵심 도구입니다.
import ipywidgets as widgets  # 슬라이더, 버튼 등 대화형 UI 구성 요소를 만드는 라이브러리입니다.
from ipywidgets import interact  # 슬라이더 조작에 따라 특정 함수를 즉시 다시 실행시켜주는 인터랙티브 데코레이터 함수입니다.

# 주피터 노트북 및 matplotlib 시각화 환경에서 OS별 한글 폰트가 깨지는 현상을 방지하기 위한 초기 설정입니다.
if platform.system() == "Darwin":  # 사용자의 OS가 macOS(Darwin)일 경우
    plt.rcParams["font.family"] = "AppleGothic"  # 애플 운영체제 기본 한글 폰트인 AppleGothic을 적용합니다.
elif platform.system() == "Windows":  # 사용자의 OS가 Windows일 경우
    plt.rcParams["font.family"] = "Malgun Gothic"  # 마이크로소프트 기본 한글 폰트인 맑은 고딕을 적용합니다.

# 그래프의 축 좌표 범위 등에 마이너스(-) 기호가 사각형 등으로 깨지는 에러를 방지합니다.
plt.rcParams["axes.unicode_minus"] = False
```

### 📈 2단계: 수학적 데이터 생성 및 그리기 함수 정의 (Data Generation & Functions)
고정비, 단위 변동비 비율(%), 판매 가격을 매개변수로 하여 손익분기 판매량을 도출하고, 이익 구간(녹색 빗금 영역)과 손실 구간(적색 빗금 영역)을 한 화면에 렌더링하는 `plot_break_even` 함수를 정의합니다.

```python
def plot_break_even(fixed_cost, var_cost_ratio, price):
    # fixed_cost: 고정비 (단위: 만 원)
    # var_cost_ratio: 판매 단가 대비 변동비의 비율 (%)
    # price: 단위 판매 단가 (SaaS 월 구독 요금 등, 단위: 만 원)
    
    # 단위당 변동비(Variable Cost per unit)를 가격의 비율로부터 환산합니다.
    # 예: 판매 단가가 5만 원이고 비율이 40%이면 단위 변동비는 2만 원이 됩니다.
    unit_var_cost = price * (var_cost_ratio / 100.0)
    
    # matplotlib figure를 생성합니다. (비즈니스 대시보드 뷰)
    fig, ax = plt.subplots(figsize=(11, 6))
    
    # 판매 단가가 단위당 변동비보다 작거나 같으면 이익이 절대로 날 수 없습니다. (적자 수렁 상태)
    # 수학적으로 분모 (Price - Variable Cost)가 0 이하가 되기 때문입니다.
    if price <= unit_var_cost:
        # 이 특수한 파탄 상태에 대한 경고 메시지를 플롯 중앙에 배치하고 종료합니다.
        ax.text(0.5, 0.5, "[경고]: 판매 단가가 단위당 변동비보다 낮습니다!\n제품을 팔 때마다 손실이 발생하여 손익분기점에 도달할 수 없습니다.",
                fontsize=14, color="red", ha="center", va="center", weight="bold",
                bbox=dict(facecolor="white", alpha=0.8, edgecolor="red", boxstyle="round,pad=1"))
        ax.set_xlim(0, 100)
        ax.set_ylim(0, 100)
        ax.set_axis_off()
        plt.show()
        return

    # 손익분기점(BEP)을 이루는 판매량(x)을 계산합니다.
    # 부등식: Price * x > Fixed Cost + Variable Cost * x
    # -> (Price - Variable Cost) * x > Fixed Cost
    # -> x > Fixed Cost / (Price - Variable Cost)
    bep_units = fixed_cost / (price - unit_var_cost)
    
    # 그래프를 조밀하게 그리기 위한 x 범위(판매량)를 설정합니다.
    # 손익분기점을 시각적으로 중앙 근처에 두기 위해 최대 표시 범위를 bep_units의 약 2배 정도로 동적 계산합니다.
    # 단, bep_units가 너무 크거나 작을 때의 적정선을 위해 최소 50에서 최대 2000 사이로 범위를 제한합니다.
    max_x = int(max(50, min(2000, bep_units * 2)))
    x = np.linspace(0, max_x, 500)
    
    # 매출 함수 R(x) = Price * x 를 생성합니다.
    revenue = price * x
    # 비용 함수 C(x) = Fixed Cost + Variable Cost * x 를 생성합니다.
    cost = fixed_cost + unit_var_cost * x
    
    # 매출선과 비용선을 작도합니다.
    # Blue (#0B79D0) 계열과 Stark orange 계열을 활용하여 신뢰도 높은 경영상황실 느낌을 줍니다.
    ax.plot(x, revenue, color="#0B79D0", linewidth=3, label=f"총매출 R(x) = {price:.1f}x")
    ax.plot(x, cost, color="#E67E22", linewidth=3, label=f"총비용 C(x) = {fixed_cost:.0f} + {unit_var_cost:.1f}x")
    
    # 손익분기점(BEP) 교점 좌표를 계산합니다.
    bep_revenue = price * bep_units
    
    # 손익분기점 교점을 녹색 별 마크로 돋보이게 표시합니다.
    ax.plot(bep_units, bep_revenue, marker="*", color="#27AE60", markersize=15, zorder=10,
            label=f"손익분기점 (BEP: {bep_units:.1f}명)")
    
    # 손익분기점 위치에서 축으로 떨어지는 점선을 보조선으로 그립니다.
    ax.axvline(bep_units, color="gray", linestyle="--", linewidth=1.5)
    ax.axhline(bep_revenue, color="gray", linestyle="--", linewidth=1.5)
    
    # 이익 영역(Profit > 0)인 BEP 우측 영역을 부드러운 녹색/파란색 빗금(Hatch)으로 동적 드로잉합니다.
    # fill_between 함수를 활용하여 매출선이 비용선보다 높은 위쪽 공간을 메웁니다.
    ax.fill_between(x, revenue, cost, where=(x >= bep_units), interpolate=True,
                    facecolor="#2ECC71", alpha=0.25, hatch="//", edgecolor="#27AE60",
                    label="이익 영역 (Profit > 0)")
                    
    # 손실 영역(Profit < 0)인 BEP 좌측 영역을 부드러운 빨간색 빗금으로 동적 드로잉합니다.
    # 비용선이 매출선보다 높은 위쪽 공간을 메웁니다.
    ax.fill_between(x, cost, revenue, where=(x < bep_units), interpolate=True,
                    facecolor="#E74C3C", alpha=0.15, hatch="\\\\", edgecolor="#C0392B",
                    label="손실 영역 (Profit < 0)")
                    
    # 그래프 내 텍스트 주석으로 현재 손익분기점 정보를 표시합니다.
    ax.text(bep_units + (max_x * 0.03), bep_revenue - (bep_revenue * 0.05),
            f"BEP: {bep_units:.1f} 명 판매 시 돌파\n(매출 {bep_revenue:.1f}만 원)",
            color="#27AE60", fontsize=11, weight="bold",
            bbox=dict(facecolor="white", alpha=0.9, edgecolor="#27AE60", boxstyle="round,pad=0.5"))
    # 고정비 라인(Fixed Cost)을 수평선으로 가볍게 그려주어 비용 구조의 하한선을 인지시킵니다.
    ax.axhline(fixed_cost, color="red", linestyle=":", alpha=0.5, label=f"고정비 (F) = {fixed_cost}만 원")

    # 차트의 가독성을 최상으로 확보하기 위해 축 레이블과 제목을 설정합니다.
    ax.set_title("SaaS 비즈니스 손익분기점(BEP) 분석 및 최적화 시뮬레이터", fontsize=15, pad=20, weight="bold")
    ax.set_xlabel("사용자 수 / 판매량 (명)", fontsize=12, labelpad=10)
    ax.set_ylabel("금액 (단위: 만 원)", fontsize=12, labelpad=10)
    
    # x축 and y축의 최소값을 0으로 고정하여 현실적인 경제 지표 뷰를 만듭니다.
    ax.set_xlim(0, max_x)
    ax.set_ylim(0, bep_revenue * 2)
    
    # 그리드 격자를 엷게 적용합니다.
    ax.grid(True, linestyle=":", alpha=0.6)
    
    # 범례를 좌상단에 고급스럽게 배치합니다.
    ax.legend(loc="upper left", frameon=True, facecolor="white", shadow=True, fontsize=10)
    
    plt.tight_layout()
    plt.show()
```

### 🎛️ 3단계: 인터랙티브 대시보드 가동 및 분석 (Plotting & Execution)
아래 코드를 실행하여 최적화 대시보드를 구동합니다. 슬라이더를 통해 고정비($F$), 단위 변동비 비율, 판매 가격($P$)을 조작할 때마다 흑자/적자 영역과 손익분기점이 실시간으로 연동되는 것을 관찰해 보세요. 
예를 들어, 서버 최적화를 통해 고정비(Fixed Cost)를 낮추거나, 가격을 소폭 인상했을 때 비즈니스 생존 경계선(BEP)이 어떻게 당겨지는지 직접 체감할 수 있습니다.

```python
# interact 함수를 호출하여 비즈니스 매개변수 위젯을 연동시킵니다.
# 고정비(만 원), 변동비 비율(%), 구독가(만 원)를 슬라이더로 조작할 수 있게 합니다.
interact(
    plot_break_even,
    fixed_cost=widgets.IntSlider(value=300, min=50, max=1000, step=50, description="고정비(만)"),
    var_cost_ratio=widgets.IntSlider(value=20, min=0, max=90, step=5, description="변동비율(%)"),
    price=widgets.FloatSlider(value=2.0, min=0.5, max=10.0, step=0.5, description="구독요금(만)")
)
```
