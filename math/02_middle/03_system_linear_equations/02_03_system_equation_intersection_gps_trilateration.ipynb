# 🟢 3단계: 연립일차방정식 (System of Linear Equations)
> **서로 다른 두 직선의 아름다운 만남, 그리고 현대 문명의 공간적 눈이 된 삼각측량**

---

### 1. 묵상과 사유 (철학적·종교적 관점)
하나의 일차방정식이가 광활한 우주 위에 홀로 뻗어 나가는 무한한 가능성의 한 직선이라면, 연립방정식(System of Equations)은 서로 다른 배경과 기울기를 가진 두 직선이 만나 타협을 이루는 **'조화와 합일(Harmony & Unity)'**의 우주입니다.

* **서로 다른 두 가치의 만남과 평화의 한 점**
  각기 다른 조건식($ax + by = c$ 와 $dx + ey = f$)은 세상의 다양한 이해관계나 서로 다른 신념의 궤적을 닮았습니다. 이 두 궤적이 한 평면 위에서 교차할 때, 오직 단 하나의 공유점(Intersection)만이 두 식을 동시에 만족하는 해(Solution)가 됩니다. 
  이는 인생과 공동체 속에서 서로 다른 주장과 삶의 방식(기울기)을 가진 존재들이, 각자의 개성을 상실하지 않으면서도 서로를 인정하고 조화를 이루며 찾아내는 유일한 평화의 접점(교점)을 상기시킵니다.

* **평행(Parallel)과 일치(Coincidence)의 경계**
  기울기가 같아 끝내 만나지 못하는 두 직선(평행)은 서로 완전히 마음을 닫고 마주 보지 못하는 차가운 평행선을 보여줍니다. 이 상태는 관계 속에서의 '단절과 무해(해 없음)'를 뜻합니다.
  반면 두 직선이 완전히 겹쳐 구분할 수 없는 상태(일치)는 개성 없이 획일화되어 서로 동질화된 '개성 상실과 경계의 부재(해의 과잉)'를 떠올리게 합니다. 
  서로 다른 기울기를 유지한 채 하나의 건강한 점(교점)에서 마주하고, 그 한 점을 지나 각자의 길을 의연히 걸어가는 기하학적 형태는 관계 속의 진정한 존중을 가르쳐 줍니다.

* **소거(Elimination)의 지혜: 단순함으로 본질에 도달하기**
  연립방정식을 풀 때 가감법이나 대입법을 통해 변수 하나를 제거(소거)하여 $x$나 $y$의 단일 본질에 도달하듯, 인생의 복잡하고 얽힌 문제들도 무수한 잡음(Noise) 변수들을 덜어내고 본질적인 핵심 변수 하나만을 직시할 때 비로소 매듭이 풀리기 시작함을 상징합니다.

### 2. 왜 사용하는가? 실제 생활에서의 적용점

* **현대 모빌리티 문명의 눈: GPS 위성 삼각측량**
  우리가 스마트폰 지도 앱을 켜고 길을 찾을 때, 스마트폰은 우주에 떠 있는 위성 3~4개로부터 신호를 받아 위성과의 거리에 관한 연립방정식을 풉니다. 위성을 중심으로 그려지는 여러 구(Sphere)들의 방정식들이 연립되어 교차하는 단 하나의 위도와 경도 좌표(내 위치)를 나노초 단위로 정확히 계산해 냅니다. 연립방정식이 현대 문명의 공간적 눈이 되어 줍니다.

* **시장의 평화를 결정하는 보이지 않는 손: 수요와 공급의 균형 가격**
  경제학에서 시장의 가격과 거래량은 소비자의 구매 욕구를 담은 '수요 곡선식'과 공급자의 생산 의지를 담은 '공급 곡선식'이 만나는 교점에서 최종 결정됩니다. 이 연립방정식의 해가 바로 시장의 혼란을 막고 거래를 안정시키는 '균형 가격(Market Equilibrium)'입니다.

* **가상 세계의 물리적 충돌 감지 (Collision Detection)**
  3D 레이싱 게임이나 메타버스 공간에서 자동차가 벽에 부딪히거나 캐릭터가 장애물을 밟는 순간을 실시간 연산하는 것은, 움직이는 물체들의 이동 궤적 방정식을 연립하여 공통의 교점(충돌 좌표)이 발생하는지 실시간으로 감시하고 튕겨 나가게 하는 대수적 제어 장치입니다.

## 1. 일차연립방정식 교점 시각화 (System Equation Intersection)

아래 실습은 좌표평면 위에 두 일차방정식 $ax + by = c$와 $dx + ey = f$가 그리는 직선들을 렌더링하고, 사용자가 슬라이더로 계수를 조정함에 따라 교점(해)의 위치가 실시간으로 부드럽게 이동하는 인터랙티브 그래프 대시보드입니다.
기울기가 같아 끝내 만나지 못하는 '해 없음(평행)' 상태와, 두 직선이 완벽히 겹쳐져 선 그 자체가 해가 되는 '해 무한(일치)' 상태의 기하학적 순간을 직접 계수를 맞춰 관찰해 보세요.

### 🛠️ 1단계: 도구 준비 (Imports & Settings)
실습을 진행하기 위해 필요한 파이썬 라이브러리(`numpy`, `matplotlib.pyplot`, `ipywidgets` 등)를 불러오고, 그래프에서 한글 폰트가 깨지지 않도록 한글 환경 설정을 수행합니다.

```python
import platform
import numpy as np  # 대규모 다차원 배열과 행렬 연산을 위한 라이브러리입니다.
import matplotlib.pyplot as plt  # 데이터를 시각화(그래프 그리기)하기 위한 핵심 도구입니다.
import ipywidgets as widgets  # 슬라이더, 버튼 등 대화형 UI 구성 요소를 만드는 라이브러리입니다.
from ipywidgets import interact  # 슬라이더 조작에 따라 특정 함수를 즉시 다시 실행시켜주는 인터랙티브 데코레이터 함수입니다.

# 주피터 노트북 및 matplotlib 시각화 환경에서 OS별 한글 폰트가 깨지는 현상을 방지하기 위한 초기 설정입니다.
if platform.system() == "Darwin":  # 사용자의 OS가 macOS(Darwin)일 경우
    plt.rcParams["font.family"] = "AppleGothic"  # 애플 운영체제 기본 한글 폰트인 AppleGothic을 적용합니다.
elif platform.system() == "Windows":  # 사용자의 OS가 Windows일 경우
    plt.rcParams["font.family"] = "Malgun Gothic"  # 마이크로소프트 기본 한글 폰트인 맑은 고딕을 적용합니다.

# 그래프의 축 좌표 범위 등에 마이너스(-) 기호가 사각형 등으로 깨지는 에러를 방지합니다.
plt.rcParams["axes.unicode_minus"] = False
```

### 📈 2단계: 수학적 데이터 생성 및 그리기 함수 정의 (Data Generation & Functions)
두 일차방정식의 계수를 매개변수로 받아 각 직선의 수치 데이터를 구하고, 소거법/행렬 연산의 기본인 판별식 $D = ae - bd$를 이용해 유일한 교점 유무를 동적으로 감정하여 그래프 상에 렌더링하는 `plot_system_equations` 함수를 정의합니다.
이모지 폰트 경고 및 깨짐을 방지하기 위해 matplotlib 텍스트 출력 문구에서 특수 기호를 배제합니다.

```python
def plot_system_equations(a, b, c, d, e, f):
    # x축의 그리기 범위를 -10부터 10까지 400개의 점으로 세밀하게 설정합니다.
    x = np.linspace(-10, 10, 400)
    
    # matplotlib figure를 생성합니다.
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # 판별식 D = ae - bd 를 계산하여 연립일차방정식 해의 개수를 검정합니다.
    det = a * e - b * d
    
    # 두 직선이 유일한 교점(해)을 가지는 경우
    if det != 0:
        # 크래머 공식(Cramer's Rule)을 적용하여 교점 좌표 (x, y)를 핀포인트로 역산합니다.
        x_sol = (c * e - b * f) / det
        y_sol = (a * f - c * d) / det
        solution_text = f"교점(해): x = {x_sol:.2f}, y = {y_sol:.2f}"
    else:
        # det = 0일 때는 두 직선이 평행하거나 완전히 일치하는 경우입니다.
        # 두 직선의 계수와 상수항의 비율을 비교하여 평행(해 없음)과 일치(해 무한)를 판별합니다.
        # 0 나누기 오류를 차단하기 위해 외적 연산(Cross Multiplication) 형태로 동일성 여부를 검정합니다.
        is_coincident = (a * f == c * d) and (b * f == c * e)
        if is_coincident:
            solution_text = "[일치] 두 직선이 일치하여 해가 무한히 많습니다."
        else:
            solution_text = "[평행] 두 직선이 평행하여 해가 존재하지 않습니다."

    # 첫 번째 직선 ax + by = c 그리기
    # y의 계수인 b가 0인 경우, 직선은 y축과 평행한 수직선 x = c/a 가 됩니다.
    if b == 0:
        if a != 0:
            ax.axvline(x=c/a, color="#0B79D0", linewidth=2.5, label=f"{a}x = {c}")
    else:
        # b가 0이 아니면 정상적으로 y = (c - ax) / b 형태의 함수 데이터를 생성해 그립니다.
        y1 = (c - a * x) / b
        ax.plot(x, y1, color="#0B79D0", linewidth=2.5, label=f"{a}x + {b:+}y = {c}")

    # 두 번째 직선 dx + ey = f 그리기
    # e가 0인 경우, 직선은 y축과 평행한 수직선 x = f/d 가 됩니다.
    if e == 0:
        if d != 0:
            ax.axvline(x=f/d, color="#E67E22", linewidth=2.5, label=f"{d}x = {f}")
    else:
        # e가 0이 아니면 정상적으로 y = (f - dx) / e 형태의 함수 데이터를 생성해 그립니다.
        y2 = (f - d * x) / e
        ax.plot(x, y2, color="#E67E22", linewidth=2.5, label=f"{d}x + {e:+}y = {f}")

    # 유일한 교점이 존재할 경우, 그래프 위에 교점을 빨간색 점으로 하이라이트 표시합니다.
    if det != 0:
        # 교점 좌표 범위가 화면에 포함될 때만 그립니다.
        if -15 <= x_sol <= 15 and -15 <= y_sol <= 15:
            # macOS 환경(AppleGothic) 등에서 bold 폰트 누락 오류 경고(Failed to find font weight bold)를 방지하기 위해 weight 속성을 제거하거나 normal로 지정합니다.
            ax.plot(x_sol, y_sol, marker="o", color="red", markersize=10, zorder=5, label="연립방정식의 해 (교점)")
            ax.annotate(f"({x_sol:.1f}, {y_sol:.1f})", xy=(x_sol, y_sol), xytext=(x_sol + 0.5, y_sol + 0.5),
                        fontsize=11, weight="normal", color="red",
                        bbox=dict(facecolor="white", alpha=0.8, edgecolor="red", boxstyle="round,pad=0.3"))

    # 좌표평면 기하 뷰를 명확히 다듬기 위해 x축, y축 기준선을 회색선으로 그립니다.
    ax.axhline(0, color="black", linewidth=1)
    ax.axvline(0, color="black", linewidth=1)
    
    # 그래프 축 범위 및 그리드 레이아웃 설정
    ax.set_xlim(-10, 10)
    ax.set_ylim(-10, 10)
    ax.set_xticks(range(-10, 11, 2))
    ax.set_yticks(range(-10, 11, 2))
    ax.grid(True, linestyle=":", alpha=0.6)

    # 타이틀에 연립방정식 해 정보를 출력하고 축 레이블을 기재합니다. weight 속성을 normal로 주어 폰트 누락 오류를 차단합니다.
    ax.set_title(f"[분석 결과] {solution_text}", fontsize=13, pad=15, weight="normal")
    ax.set_xlabel("X 축", fontsize=11)
    ax.set_ylabel("Y 축", fontsize=11)
    
    # 범례 설정
    ax.legend(loc="upper right", frameon=True, shadow=True, fontsize=10)
    plt.tight_layout()
    plt.show()
```

앞서 작성한 그리기 함수에 슬라이더들을 매핑하여 작동시킵니다. 
두 직선의 기울기와 상수들을 마우스로 움직여가며, 두 직선이 하나의 선명한 교점에서 만나는 상태와 평행하게 달리는 상태, 그리고 서로 포개어지는 순간을 눈으로 추적해 보세요.

```python
# interact 함수를 연결하여 두 직선 ax + by = c, dx + ey = f 의 계수들을 조정하는 대시보드를 띄웁니다.
interact(
    plot_system_equations,
    a=widgets.FloatSlider(value=1.0, min=-5.0, max=5.0, step=0.5, description="직선1 a"),
    b=widgets.FloatSlider(value=1.0, min=-5.0, max=5.0, step=0.5, description="직선1 b"),
    c=widgets.FloatSlider(value=2.0, min=-10.0, max=10.0, step=0.5, description="직선1 c"),
    d=widgets.FloatSlider(value=1.0, min=-5.0, max=5.0, step=0.5, description="직선2 d"),
    e=widgets.FloatSlider(value=-1.0, min=-5.0, max=5.0, step=0.5, description="직선2 e"),
    f=widgets.FloatSlider(value=4.0, min=-10.0, max=10.0, step=0.5, description="직선2 f")
)
```

## 2. GPS 위성 삼각측량 시뮬레이터 (GPS Trilateration)

우리가 매일 사용하는 스마트폰 내비게이션이나 배달 앱의 근간에는 **'삼각측량(Trilateration)'**이라는 기하학적 연립방정식이 숨어 있습니다.
2차원 평면 공간 상에 위치를 정확히 알고 있는 3개의 위성 기지국이 있고, 각 위성에서 전파 도달 시간(ToA, Time of Arrival)을 통해 단말기(Target)까지의 실측 거리 $d_i$를 구했다고 가정해 봅시다.
단말기의 좌표 $(x, y)$는 위성들을 중심으로 하고 $d_i$를 반지름으로 하는 세 원의 방정식을 연립할 때, 그 **공통 교점**으로 핀포인트 계산해 낼 수 있습니다.

### 📐 수학적 모델링과 선형화(Linearization)의 묘미

3개 위성의 고정 좌표를 각각 $A(x_a, y_a)$, $B(x_b, y_b)$, $C(x_c, y_c)$라 하고, 단말기의 위치를 $T(x, y)$라 합시다.
이때 단말기로부터 세 위성까지의 거리에 관한 연립방정식은 다음과 같이 **2차 곡선의 방정식**들로 표현됩니다.

1. $(x - x_a)^2 + (y - y_a)^2 = d_a^2$
2. $(x - x_b)^2 + (y - y_b)^2 = d_b^2$
3. $(x - x_c)^2 + (y - y_c)^2 = d_c^2$

이것은 미지수 $x, y$에 대한 비선형(Non-linear) 2차 연립방정식으로, 직접 풀기에는 다소 복잡해 보입니다. 
하지만 연립방정식의 본질인 **'소거(Elimination)'의 지혜**를 발휘하면, 놀랍도록 단순한 **일차방정식**으로 변환할 수 있습니다.

#### 1) 2차 항($x^2, y^2$) 소거하기
식 (1), (2), (3)을 각각 제곱하여 전개해 보면 공통적으로 $x^2 + y^2$ 항이 나타납니다.
- 식(1) 전개: $x^2 - 2x_a x + x_a^2 + y^2 - 2y_a y + y_a^2 = d_a^2$
- 식(2) 전개: $x^2 - 2x_b x + x_b^2 + y^2 - 2y_b y + y_b^2 = d_b^2$

여기서 **식(1)에서 식(2)를 빼주면(가감법)**, 2차항인 $x^2$과 $y^2$이 통째로 날아가며 다음과 같은 **일차방정식(직선의 방정식)**이 유도됩니다.
$$2(x_b - x_a)x + 2(y_b - y_a)y = d_a^2 - d_b^2 - (x_a^2 - x_b^2) - (y_a^2 - y_b^2)$$
이 직선을 기하학적으로 **'근축(Radical Axis)'**이라 부르며, 두 원 $A$와 $B$가 만나는 교점들을 지나는 직선이 됩니다.

#### 2) 연립일차방정식 완성하기
마찬가지로 **식(1)에서 식(3)을 빼주면** 또 하나의 일차방정식이 유도됩니다.
$$2(x_c - x_a)x + 2(y_c - y_a)y = d_a^2 - d_c^2 - (x_a^2 - x_c^2) - (y_a^2 - y_c^2)$$

결국 세 원의 복잡한 2차 연립방정식이 다음과 같은 형태의 **두 일차방정식의 연립(선형 연립방정식)**으로 단순화됩니다!
$$\begin{cases} K_1 x + K_2 y = K_3 \\ K_4 x + K_5 y = K_6 \end{cases}$$
(여기서 $K_1 \sim K_6$은 위성의 좌표와 거리 실측값으로 채워지는 기지의 상수들입니다.)

이 두 직선의 교점을 가감법이나 대입법으로 구하면, 세 원이 만나는 단 하나의 공통 교점이자 단말기의 최종 위치 $T(x, y)$가 핀포인트로 정확히 계산됩니다.
우리가 중학교에서 배우는 '연립일차방정식의 소거' 원리가 현대 모빌리티 문명의 근간인 GPS의 핵심 알고리즘으로 작동하고 있는 것입니다.

### 🛠️ 1단계: 도구 준비 (Imports & Settings)
각 파트별 독립 실행성 보장 원칙에 맞춰, 필요한 임포트와 한글 설정을 완전히 내장하여 노트북이 항상 독립적이고 주체적으로 구동할 수 있게 조치합니다.

```python
import platform
import numpy as np  # 대규모 다차원 배열과 행렬 연산을 위한 라이브러리입니다.
import matplotlib.pyplot as plt  # 데이터를 시각화(그래프 그리기)하기 위한 핵심 도구입니다.
import ipywidgets as widgets  # 슬라이더, 버튼 등 대화형 UI 구성 요소를 만드는 라이브러리입니다.
from ipywidgets import interact  # 슬라이더 조작에 따라 특정 함수를 즉시 다시 실행시켜주는 인터랙티브 데코레이터 함수입니다.

# 주피터 노트북 및 matplotlib 시각화 환경에서 OS별 한글 폰트가 깨지는 현상을 방지하기 위한 초기 설정입니다.
if platform.system() == "Darwin":  # 사용자의 OS가 macOS(Darwin)일 경우
    plt.rcParams["font.family"] = "AppleGothic"  # 애플 운영체제 기본 한글 폰트인 AppleGothic을 적용합니다.
elif platform.system() == "Windows":  # 사용자의 OS가 Windows일 경우
    plt.rcParams["font.family"] = "Malgun Gothic"  # 마이크로소프트 기본 한글 폰트인 맑은 고딕을 적용합니다.

# 그래프의 축 좌표 범위 등에 마이너스(-) 기호가 사각형 등으로 깨지는 에러를 방지합니다.
plt.rcParams["axes.unicode_minus"] = False
```

### 📈 2단계: 수학적 데이터 생성 및 그리기 함수 정의 (Data Generation & Functions)
세 위성의 중심 고정 좌표와 단말기의 동적 좌표 $(x_t, y_t)$를 이용해 원들을 생성하고, 원들이 단말기 한 점으로 완벽하게 수렴하여 교차하는 삼각측량 기하학을 표기하는 `plot_gps_trilateration` 함수를 정의합니다.

```python
def plot_gps_trilateration(target_x, target_y):
    # 3개 위성 기지국의 고정 평면 좌표를 설정합니다.
    # 삼각형 형태를 이루어 위치 측위의 신뢰성을 넓히는 레이아웃입니다.
    sat_a = (0.0, 8.0)    # 위성 A: Y축 상단
    sat_b = (-7.0, -4.0)  # 위성 B: 좌측 하단
    sat_c = (7.0, -4.0)   # 위성 C: 우측 하단
    
    # matplotlib figure를 생성합니다. 공간 좌표이므로 9 x 9 정비율로 빌드합니다.
    fig, ax = plt.subplots(figsize=(9, 9))
    
    # 타겟(단말기) 좌표로부터 각 위성 기지국까지의 기하학적 유클리드 거리를 역산합니다.
    # 이 거리가 각 위성이 스마트폰 신호 분석을 통해 알아낸 실제 전파 도달 거리(반지름 d)가 됩니다.
    dist_a = np.sqrt((target_x - sat_a[0])**2 + (target_y - sat_a[1])**2)
    dist_b = np.sqrt((target_x - sat_b[0])**2 + (target_y - sat_b[1])**2)
    dist_c = np.sqrt((target_x - sat_c[0])**2 + (target_y - sat_c[1])**2)
    
    # 각 위성 위치를 차트 상에 파란색 삼각형 마커로 표기합니다.
    ax.plot(sat_a[0], sat_a[1], "^", color="#0B79D0", markersize=12, zorder=5, label="위성 A (Satellite A)")
    ax.plot(sat_b[0], sat_b[1], "^", color="#2FA85D", markersize=12, zorder=5, label="위성 B (Satellite B)")
    ax.plot(sat_c[0], sat_c[1], "^", color="#E67E22", markersize=12, zorder=5, label="위성 C (Satellite C)")
    
    # 위성 라벨링을 추가합니다. weight를 normal로 변경해 폰트 깨짐 경고를 방지합니다.
    ax.text(sat_a[0], sat_a[1] + 0.6, f"A (0, 8)\\n거리={dist_a:.1f}", ha="center", fontsize=10, weight="normal")
    ax.text(sat_b[0] - 0.6, sat_b[1] - 0.6, f"B (-7, -4)\\n거리={dist_b:.1f}", ha="right", fontsize=10, weight="normal")
    ax.text(sat_c[0] + 0.6, sat_c[1] - 0.6, f"C (7, -4)\\n거리={dist_c:.1f}", ha="left", fontsize=10, weight="normal")

    # 각 위성으로부터 전파 도달 거리 d를 반지름으로 하는 원(Sphere의 2D 평면 투영)을 그립니다.
    # 세 원의 반지름 영역을 서로 다른 투명 컬러 패치로 렌더링합니다.
    circle_a = plt.Circle(sat_a, dist_a, color="#0B79D0", fill=False, linestyle="-", linewidth=2.0, alpha=0.7)
    circle_b = plt.Circle(sat_b, dist_b, color="#2FA85D", fill=False, linestyle="-", linewidth=2.0, alpha=0.7)
    circle_c = plt.Circle(sat_c, dist_c, color="#E67E22", fill=False, linestyle="-", linewidth=2.0, alpha=0.7)
    
    # ax 패치에 추가하여 차트 위에 원들을 얹어줍니다.
    ax.add_patch(circle_a)
    ax.add_patch(circle_b)
    ax.add_patch(circle_c)
    
    # 세 원의 영역 내부에 엷은 반투명 채우기를 적용하여 삼각측량 중첩 영역을 강조합니다.
    circle_fill_a = plt.Circle(sat_a, dist_a, color="#0B79D0", fill=True, alpha=0.04)
    circle_fill_b = plt.Circle(sat_b, dist_b, color="#2FA85D", fill=True, alpha=0.04)
    circle_fill_c = plt.Circle(sat_c, dist_c, color="#E67E22", fill=True, alpha=0.04)
    ax.add_patch(circle_fill_a)
    ax.add_patch(circle_fill_b)
    ax.add_patch(circle_fill_c)

    # 세 원이 연립하여 완벽하게 만나는 교차점인 단말기(Target)의 위치를 하이라이트합니다.
    # 빨간색 별표 기호로 시인성을 확보합니다. weight를 normal로 설정하여 글리프 경고를 제거합니다.
    ax.plot(target_x, target_y, "*", color="red", markersize=16, zorder=10, label="수신 단말기 (Target)")
    ax.annotate(f"단말기 위치: ({target_x:.2f}, {target_y:.2f})", xy=(target_x, target_y),
                xytext=(target_x + 0.8, target_y + 0.8), fontsize=11, weight="normal", color="red",
                bbox=dict(facecolor="white", alpha=0.9, edgecolor="red", boxstyle="round,pad=0.4"))
    
    # 기준이 되는 x축과 y축 회색 가이드 실선
    ax.axhline(0, color="gray", linestyle=":", alpha=0.5)
    ax.axvline(0, color="gray", linestyle=":", alpha=0.5)
    
    # 그래프 축 범위 고정 및 종횡비 일치 설정
    # 원이 찌그러진 타원으로 보이지 않도록 'equal' 종횡비를 인가하는 것이 시각화 신뢰도 면에서 매우 중요합니다.
    ax.set_aspect('equal')
    ax.set_xlim(-15, 15)
    ax.set_ylim(-15, 15)
    ax.set_xticks(range(-15, 16, 3))
    ax.set_yticks(range(-15, 16, 3))
    ax.grid(True, linestyle="--", alpha=0.4)
    
    # 축 제목 및 차트 명칭 명기. weight를 normal로 줍니다.
    ax.set_title("[GPS 삼각측량] 3개 위성 신호의 원 방정식 연립과 해 공간 수렴", fontsize=13, pad=15, weight="normal")
    ax.set_xlabel("X 좌표 (km)", fontsize=11)
    ax.set_ylabel("Y 좌표 (km)", fontsize=11)
    
    # 범례 표기
    ax.legend(loc="upper left", frameon=True, shadow=True, fontsize=9)
    plt.tight_layout()
    plt.show()
```

### 🎛️ 3단계: 시뮬레이션 가동 및 흐름 관찰 (Plotting & Execution)
슬라이더를 조작하여 가상의 단말기(Target) 좌표를 움직여 봅니다. 
좌표가 바뀔 때마다 위성들과의 거리가 실시간으로 계산되며, 세 원이 정확히 타겟 점 하나를 정밀 조준하며 수렴하는 마술 같은 삼각측량 기하 구조를 관찰해 보세요. 
우리가 일상에서 지도를 켜고 걷는 매 순간마다, 이 3개 원의 연립방정식이 나노초 단위로 휴대폰 AP 프로세서 내부에서 연산되고 있습니다.

```python
# interact 함수를 연결해 타겟의 target_x 및 target_y 좌표를 조작하는 위젯 대시보드를 띄웁니다.
# -10부터 10까지 0.5 간격으로 자유롭게 가상 위치를 설정할 수 있습니다.
interact(
    plot_gps_trilateration,
    target_x=widgets.FloatSlider(value=3.0, min=-10.0, max=10.0, step=0.5, description="단말기 X"),
    target_y=widgets.FloatSlider(value=2.0, min=-10.0, max=10.0, step=0.5, description="단말기 Y")
)
```
