### 1. 묵상과 사유 (철학적·종교적 관점)
기하학(Geometry)은 만물의 형태와 구조 속에 깃든 신성한 질서와 비례를 탐구하는 학문입니다. 점, 선, 면의 단순한 요소들이 모여 우주와 물질을 지탱하는 완벽한 역학적 평형을 완성합니다.

* **수직과 평행: 사랑과 정의가 교차하는 영적 뼈대**
  지면과 정밀한 직각을 이루는 수직선(Vertical Line)은 중력의 법칙 앞에 한 치의 타협도 없는 공평한 **'공의(Justice)'**를 상징합니다. 반면, 영원히 만나지 않고 나란히 뻗어가는 평행선(Parallel Lines)은 동등한 거리에서 서로를 존중하며 동행하는 **'평등과 동반(사랑)'**을 상징합니다. 
  이 두 축(수직과 평행)이 만날 때 세상은 무너지지 않는 굳건한 기초(Foundation)를 얻게 되며, 이는 동서고금의 성스러운 건축물들이 기하학적 대칭을 통해 신의 정의와 사랑을 표현하고자 했던 건축 철학과 긴밀히 닿아 있습니다.

* **투영(Projection)의 비유: 보이지 않는 본질의 2차원 그림자**
  3차원 공간 속 입체 도형이 카메라 관점에 따라 2차원 평면 위에 찌그러진 그림자 형태로 투영되는 현상은 깊은 사유를 낳습니다. 우리가 발을 딛고 살아가는 물리적 세계와 현상들은, 어쩌면 더 높은 차원에 존재하는 영원하고 불변하는 이데아(Idea)나 영적 본질이 이 지상이라는 낮은 평면 위에 드리운 **'그림자(투영)'**일지 모릅니다. 
  그림자의 모양(2D 투영선)은 빛의 각도와 위치에 따라 매 순간 비틀리고 변화하지만, 그 그림자를 만드는 고차원 본체(3D 입체)의 대수적 기하 규칙은 불변합니다. 기하학을 배운다는 것은 변화무쌍한 현상 너머의 영원한 규칙을 인지하는 영안을 뜨는 과정입니다.

### 2. 왜 사용하는가? 실제 생활에서의 적용점

* **인간의 문명을 안전하게 지탱하는 방패: 토목 및 건축 설계**
  우리가 매일 건너는 교량, 초고층 빌딩, 터널 등의 대형 구조물들은 자칫 균형이 깨지면 중력의 하중을 이기지 못하고 붕괴하게 됩니다. 공학자들은 기하학의 평행성과 수직성의 공식을 정교하게 분석하여, 외부의 힘과 진동이 구조물의 수직 기둥을 타고 땅으로 고르게 분산되도록 설계합니다. 기하는 생명을 지키는 역학적 설계의 근본입니다.

* **초정밀 설계 도면의 뼈대: CAD(Computer-Aided Design)**
  스마트폰의 조밀한 회로 기판부터 거대한 항공기 동체에 이르기까지, 정밀 조립을 요하는 모든 현대 기계 부품은 수치 제어 기하학 도면을 바탕으로 깎아냅니다. 선들의 사잇각, 면의 평행도, 곡률 반경의 대수적 연산은 공장에 정확한 가공 명령을 내리는 공용 기하학 언어입니다.

* **3차원 가상세계를 2D 모니터에 구현하는 기술: 3D 그래픽 투영**
  우리가 즐기는 3D 모바일 게임이나 메타버스, VR 헤드셋 화면의 본질은 사실 '원근 투영(Perspective Projection)' 계산기입니다. 가상 공간에 정의된 3D 기하 데이터(꼭짓점들의 x, y, z 좌표)를 플레이어 카메라의 실시간 회전각과 원근 공식에 따라 모니터 화면의 2D 화소 좌표로 빠르게 번역하여 실감 나는 입체감을 전달합니다.

## 1. 기하 작도 샌드박스 (Geometry Sandbox)

아래 실습은 좌표평면 상에 기준이 되는 직선 $L_1$(점 A와 점 B를 지나는 직선)을 정의하고,
제3의 점 $C$를 지나면서 $L_1$에 완벽하게 수직(직각)이거나 평행한 직선을 대수적으로 계산하여 실시간으로 좌표평면 위에 그려보는 샌드박스입니다.
기울기의 곱이 $-1$이 되는 수직의 법칙과, 기울기가 완전히 같아지는 평행의 원리를 눈으로 확인해 보세요.

### 🛠️ 1단계: 도구 준비 (Imports & Settings)
기하학 도형과 샌드박스 좌표계를 구현하기 위해 필요한 파이썬 라이브러리를 임포트하고, 플롯 내에서 한글 깨짐 및 마이너스 기호 깨짐 오류를 방지하기 위해 환경 초기화를 수행합니다.

```python
import platform  # 운영체제 정보를 읽어와 한글 글꼴 설정을 대응하기 위해 가져옵니다.
import numpy as np  # 직선 구간 생성 및 삼각함수 연산을 담당하는 라이브러리입니다.
import matplotlib.pyplot as plt  # 2D 좌표평면 기하 작도를 수행하는 시각화 도구입니다.
import ipywidgets as widgets  # 점들의 좌표 및 작도 모드 변환 슬라이더를 구성하는 위젯 패키지입니다.
from ipywidgets import interact  # 슬라이더 입력을 기하 시각화 함수에 실시간 연동해주는 가교 데코레이터입니다.

# 주피터 차트 내부에서 한글 폰트가 사각형으로 깨지는 에러를 방지하는 설정입니다.
if platform.system() == "Darwin":  # macOS의 경우
    plt.rcParams["font.family"] = "AppleGothic"  # Apple 고딕체를 설정합니다.
elif platform.system() == "Windows":  # Windows의 경우
    plt.rcParams["font.family"] = "Malgun Gothic"  # 맑은 고딕체를 설정합니다.

# 그래프 축 상에서 음수(-) 기호가 특수문자 오류로 깨져서 깨짐 현상이 발생하는 것을 방지합니다.
plt.rcParams["axes.unicode_minus"] = False
```

### 📐 2단계: 평행/수직 해석기하 연산 및 샌드박스 드로잉 함수 정의 (Data Generation & Functions)
세 점 $A, B, C$의 좌표와 설정된 관계(수직/평행/일반교차)에 의거하여 직선의 기울기들과 교점 좌표를 계산하고,
두 직선이 이루는 사잇각 $\theta = \arctan \left| \frac{m_1 - m_2}{1 + m_1 m_2} \right|$을 산출하여 시각화와 범례로 렌더링하는 `plot_geometry_sandbox` 함수를 구현합니다.

```python
def plot_geometry_sandbox(ax_x, ax_y, bx_x, bx_y, cx_x, cx_y, relation_type):
    # x축의 렌더링 도메인을 -10에서 10까지 400개 점으로 선언하여 직선 구간 연산에 활용합니다.
    x_vals = np.linspace(-10, 10, 400)
    
    # 1. 점 A와 점 B를 지나는 기준 직선 L1의 기울기 m1을 연산합니다.
    # 분모가 0이 되는 수직 직선(x의 변화량이 0)일 경우에 대한 예외 처리를 수행합니다.
    run_L1 = bx_x - ax_x
    if abs(run_L1) < 1e-5:
        m1 = 1e5  # 무한대에 가까운 매우 큰 임의의 기울기 값으로 설정
    else:
        m1 = (bx_y - ax_y) / run_L1
    
    # 기준선 L1의 방정식 y = m1 * (x - x_A) + y_A 을 정의합니다.
    # 무한대 기울기일 경우에는 수직선 형태로 그리기 위해 x 좌표를 고정 처리합니다.
    if m1 == 1e5:
        y_L1 = np.linspace(-15, 15, 400)
        x_L1 = np.ones(400) * ax_x
    else:
        y_L1 = m1 * (x_vals - ax_x) + ax_y
        x_L1 = x_vals
        
    # 2. 선택한 기하 관계(relation_type)에 따라 점 C를 지나는 직선 L2의 기울기 m2를 유도합니다.
    if relation_type == "수직 작도 (L1 ⊥ L2)":
        # 수직 조건: 두 직선의 기울기 곱이 -1이 되어야 합니다. (m1 * m2 = -1)
        # 따라서 L2의 기울기는 L1의 역수에 마이너스를 취한 값인 m2 = -1 / m1 이 됩니다.
        if abs(m1) < 1e-5:
            m2 = 1e5
        elif m1 == 1e5:
            m2 = 0.0
        else:
            m2 = -1.0 / m1
            
    elif relation_type == "평행 작도 (L1 // L2)":
        # 평행 조건: 두 직선의 기울기가 완전히 일치해야 합니다. (m2 = m1)
        m2 = m1
        
    else:  # 일반 교차 모드인 경우 점 C와 원점을 지나는 임의의 선으로 임시 매핑합니다.
        m2 = 0.5  # 고정된 임의의 기울기
        
    # 점 C를 통과하는 L2 직선의 수식 y = m2 * (x - x_C) + y_C 을 정의합니다.
    if m2 == 1e5:
        y_L2 = np.linspace(-15, 15, 400)
        x_L2 = np.ones(400) * cx_x
    else:
        y_L2 = m2 * (x_vals - cx_x) + cx_y
        x_L2 = x_vals
        
    # 3. 두 직선의 교차 각도(사잇각 theta)를 해석적으로 정교하게 계산합니다.
    # tan(theta) = abs((m1 - m2) / (1 + m1 * m2)) 공식을 활용합니다.
    if relation_type == "수직 작도 (L1 ⊥ L2)":
        angle_deg = 90.0
    elif relation_type == "평행 작도 (L1 // L2)":
        angle_deg = 0.0
    else:
        # 탄젠트 공식 및 역삼각함수(arctan) 연산을 거쳐 두 직선의 각도를 구하고 Degree 단위로 변환합니다.
        denom = 1 + m1 * m2
        if abs(denom) < 1e-5:
            angle_deg = 90.0
        else:
            tan_theta = abs((m1 - m2) / denom)
            angle_deg = np.degrees(np.arctan(tan_theta))
            
    # 시각화 Figure 준비 (8x8 대칭형 도화지, 배경색 #ffffff 흰색)
    fig, ax = plt.subplots(figsize=(8, 8), facecolor='#ffffff')
    ax.set_facecolor('#ffffff')
    
    # 1. 기준선 L1 그리기: Stark Red (#ff5a5f) 2.5px 굵은 실선으로 플로팅합니다.
    ax.plot(x_L1, y_L1, color='#ff5a5f', linewidth=2.5, label='기준선 L1 (AB)')
    
    # 2. 작도선 L2 그리기: Stark Red (#ff5a5f) 2.5px 대시(점선) 스타일로 구분되게 플로팅합니다.
    ax.plot(x_L2, y_L2, color='#ff5a5f', linestyle='--', linewidth=2.5, label='작도선 L2 (C 통과)')
    
    # 3. 주요 점 A, B, C를 강조하기 위해 Accent Green (#06d6a0) 색상 원형 마커로 찍고 테두리를 입힙니다.
    ax.plot(ax_x, ax_y, 'o', color='#06d6a0', markersize=9, markeredgecolor='black', markeredgewidth=2)
    ax.text(ax_x + 0.3, ax_y + 0.3, f'A ({ax_x}, {ax_y})', fontweight='bold')
    
    ax.plot(bx_x, bx_y, 'o', color='#06d6a0', markersize=9, markeredgecolor='black', markeredgewidth=2)
    ax.text(bx_x + 0.3, bx_y + 0.3, f'B ({bx_x}, {bx_y})', fontweight='bold')
    
    ax.plot(cx_x, cx_y, 'o', color='#06d6a0', markersize=9, markeredgecolor='black', markeredgewidth=2)
    ax.text(cx_x + 0.3, cx_y + 0.3, f'C ({cx_x}, {cx_y})', fontweight='bold')
    
    # 좌표의 격자 중심점(0,0)을 잡아주기 위해 x축과 y축의 기준선을 3px 굵은 검정 실선으로 교차시킵니다.
    ax.axhline(0, color='black', linewidth=3)
    ax.axvline(0, color='black', linewidth=3)
    
    # 보조 그리드선을 연한 회색 점선으로 깔끔하게 매칭합니다.
    ax.grid(True, color='#e0e0e0', linestyle='--', linewidth=1)
    
    # 뷰포트 스케일 고정 (가로/세로 비율을 동일하게 1:1 매칭하여 각도가 시각적으로 찌그러지지 않게 합니다.)
    ax.set_aspect('equal')
    ax.set_xlim(-10, 10)
    ax.set_ylim(-10, 10)
    
    # 범례 설정: Vibrant Yellow (#ffd166) 바탕에 2px의 두꺼운 검정 테두리 적용
    legend = ax.legend(loc='upper left', fontsize=11, frameon=True)
    frame = legend.get_frame()
    frame.set_facecolor('#ffd166')
    frame.set_edgecolor('black')
    frame.set_linewidth(2)
    
    # 차트 주변 굵은 테두리(Spines) 설정 (3px 검정)
    for spine in ax.spines.values():
        spine.set_color('black')
        spine.set_linewidth(3)
        
    # 차트 텍스트 렌더링 (사잇각 정보 동적 표시)
    ax.set_title(f'기하 샌드박스: {relation_type} (두 선의 사잇각: {angle_deg:.1f}°)', fontsize=14, fontweight='bold', pad=15)
    ax.set_xlabel('x 좌표', fontsize=11)
    ax.set_ylabel('y 좌표', fontsize=11)
    
    plt.show()
```

### 🎛️ 3단계: 기하 작도 샌드박스 실행 및 관찰 (Plotting & Execution)
슬라이더로 점 A, B, C의 좌표를 옮기면서, 작도 모드 드롭다운(수직/평행)을 전환함에 따라
수식에 의거해 수직선과 평행선이 칼같이 평면에 작도되는 기하학적 섭리를 직접 확인해 봅니다.

```python
# interact 함수를 사용하여 점 A, B, C의 X, Y 좌표 슬라이더와 관계 정의 드롭다운을 연계합니다.
# 좌표 범위는 -8.0에서 8.0까지이며, 0.5 단위로 조정할 수 있도록 설정합니다.
interact(plot_geometry_sandbox, 
         ax_x=widgets.FloatSlider(value=-5.0, min=-8.0, max=8.0, step=0.5, description='점 A (x)'),
         ax_y=widgets.FloatSlider(value=-2.0, min=-8.0, max=8.0, step=0.5, description='점 A (y)'),
         bx_x=widgets.FloatSlider(value=3.0, min=-8.0, max=8.0, step=0.5, description='점 B (x)'),
         bx_y=widgets.FloatSlider(value=4.0, min=-8.0, max=8.0, step=0.5, description='점 B (y)'),
         cx_x=widgets.FloatSlider(value=0.0, min=-8.0, max=8.0, step=0.5, description='점 C (x)'),
         cx_y=widgets.FloatSlider(value=2.0, min=-8.0, max=8.0, step=0.5, description='점 C (y)'),
         relation_type=widgets.Dropdown(options=['수직 작도 (L1 ⊥ L2)', '평행 작도 (L1 // L2)'], 
                                        value='수직 작도 (L1 ⊥ L2)', description='기하 관계'))
```

## 2. 3D 원근 투영 시뮬레이터 (Perspective Projection)

실세계의 모든 3차원 입체 도형은 관찰자(카메라)의 시선 각도와 거리 비율에 따라 2차원 화면(그림자)으로 매핑되는 기하학적 투영을 거쳐야 뇌나 스크린에 맺힙니다.
아래 코드는 3D 공간 상에 존재하는 정육면체(Cube)의 기하 모델(8개 정점 및 12개 모서리선)을 정의하고,
카메라의 회전각(Pitch, Yaw, Roll)에 맞추어 **3차원 오일러 회전 행렬 변환**을 수행한 후, **카메라 원근 투영 공식**을 이용해 2차원 평면으로 축소 렌더링하는 시뮬레이터입니다.

### 🛠️ 1단계: 도구 준비 및 초기화 (Imports & Settings)
3차원 기하 행렬 연산과 차원 투영 드로잉을 구현하기 위해 필요한 라이브러리를 임포트하고 한글 설정을 재확인하여 **실습 2의 완전한 독립 실행을 보장**합니다.

```python
import platform  # 운영체제 종류별 폰트 명세 구동을 위해 임포트합니다.
import numpy as np  # 회전 행렬 연산 및 3D 정점 좌표 벡터 계산에 사용되는 필수 패키지입니다.
import matplotlib.pyplot as plt  # 2D 평면에 투영된 최종 외곽선 형태를 렌더링하기 위한 차트 라이브러리입니다.
import ipywidgets as widgets  # 회전 각도 및 거리 슬라이더 GUI 컴포넌트를 제공하는 패키지입니다.
from ipywidgets import interact  # 슬라이더 변경 수치를 시뮬레이터 함수로 전송해주는 가교입니다.

# 한글 폰트가 깨지지 않고 깔끔하게 렌더링되도록 처리하는 OS별 글꼴 명세입니다.
if platform.system() == "Darwin":
    plt.rcParams["font.family"] = "AppleGothic"
elif platform.system() == "Windows":
    plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

### 🚀 2단계: 3D 기하 회전 및 2D 원근 사영 연산 함수 정의 (Data Generation & Functions)
3차원 공간 속 정육면체의 기본 점군을 오일러 회전 행렬 $R = R_z(\gamma) R_y(\beta) R_x(\alpha)$로 변환한 후,
카메라와의 비례 비율을 이용한 원근 투영 공식 $x' = \frac{d \cdot x}{d + z}$를 적용해 2차원 화면 좌표로 축소하는 렌더러 함수 `plot_perspective_projection`을 정의합니다.

```python
# 3D 공간 상에 위치한 정육면체(Cube)의 꼭짓점 8개의 [x, y, z] 좌표 매트릭스를 정의합니다.
# 원점을 중심으로 가로, 세로, 높이의 반경이 2인 정육면체 모델입니다.
CUBE_VERTICES = np.array([
    [-2, -2, -2],
    [ 2, -2, -2],
    [ 2,  2, -2],
    [-2,  2, -2],
    [-2, -2,  2],
    [ 2, -2,  2],
    [ 2,  2,  2],
    [-2,  2,  2]
])

# 정육면체의 외곽선을 그리기 위해 서로 쌍을 이루어 연결될 12개의 모서리선(Edge) 인덱스 구조를 정의합니다.
# 예: (0, 1)은 0번 꼭짓점과 1번 꼭짓점을 잇는 모서리입니다.
CUBE_EDGES = [
    (0, 1), (1, 2), (2, 3), (3, 0),  # 밑면 4개 모서리
    (4, 5), (5, 6), (6, 7), (7, 4),  # 윗면 4개 모서리
    (0, 4), (1, 5), (2, 6), (3, 7)   # 밑면과 윗면을 세로로 잇는 기둥 모서리 4개
]

def plot_perspective_projection(pitch_deg, yaw_deg, roll_deg, d_camera):
    # 입력받은 세 방향의 회전각(Degree)을 오일러 삼각비 행렬 연산을 위해 호도법(Radian)으로 각각 환산합니다.
    pitch = np.radians(pitch_deg)  # x축 중심 회전각 (Pitch)
    yaw = np.radians(yaw_deg)      # y축 중심 회전각 (Yaw)
    roll = np.radians(roll_deg)    # z축 중심 회전각 (Roll)
    
    # 1. 3D 회전 변환 행렬(Euler Rotation Matrix)을 각각 정의합니다.
    # Rx: Pitch 회전 행렬
    R_x = np.array([
        [1, 0, 0],
        [0, np.cos(pitch), -np.sin(pitch)],
        [0, np.sin(pitch), np.cos(pitch)]
    ])
    # Ry: Yaw 회전 행렬
    R_y = np.array([
        [np.cos(yaw), 0, np.sin(yaw)],
        [0, 1, 0],
        [-np.sin(yaw), 0, np.cos(yaw)]
    ])
    # Rz: Roll 회전 행렬
    R_z = np.array([
        [np.cos(roll), -np.sin(roll), 0],
        [np.sin(roll), np.cos(roll), 0],
        [0, 0, 1]
    ])
    
    # 세 개의 개별 회전 행렬을 곱하여 복합 회전 행렬 R = Rz * Ry * Rx 을 빌드합니다.
    R = R_z @ R_y @ R_x
    
    # 2. 정육면체의 모든 3D 꼭짓점들을 회전 행렬 R을 통해 공간 상에서 입체적으로 일괄 회전시킵니다.
    # CUBE_VERTICES의 각 좌표 행렬에 회전 행렬 R의 전치 행렬을 내적하여 고성능 3차원 위치 갱신을 수행합니다.
    rotated_vertices = CUBE_VERTICES @ R.T
    
    # 3. 3D 꼭짓점들을 카메라 원근 투영 공식(Perspective Projection)을 통해 2D 화면 좌표로 매핑합니다.
    # 공식: x_prime = d * x / (d + z), y_prime = d * y / (d + z) (단, d는 카메라와 물체 사이의 거리)
    # 분모 (d_camera + z)가 0 이하가 되어 에러가 나거나 화면 뒤쪽으로 뚫고 들어오는 가상의 시각을 물리적으로 차단합니다.
    denom = d_camera + rotated_vertices[:, 2]
    denom = np.where(denom < 0.5, 0.5, denom)  # 분모가 0.5보다 작아질 경우 최소값 0.5로 하한선 고정
    
    # 원근 투영 스케일 인자를 일괄적으로 계산하여 투영 좌표 평면 배열을 구합니다.
    proj_factor = d_camera / denom
    projected_x = rotated_vertices[:, 0] * proj_factor
    projected_y = rotated_vertices[:, 1] * proj_factor
    
    # 시각화 캔버스 생성 (8x8 대칭형 도화지, 배경색 완전 흰색 #ffffff)
    fig, ax = plt.subplots(figsize=(8, 8), facecolor='#ffffff')
    ax.set_facecolor('#ffffff')
    
    # 1. 3D 입체의 골격을 시각화하기 위해 12개 모서리(Edge) 2D 사영선들을 순회하며 그립니다.
    # 모서리선은 브랜드 메인 컬러인 Stark Red(#ff5a5f) 색상과 2.5px 실선을 채용합니다.
    for edge in CUBE_EDGES:
        p1, p2 = edge
        ax.plot(
            [projected_x[p1], projected_x[p2]], 
            [projected_y[p1], projected_y[p2]], 
            color='#ff5a5f', linewidth=2.5
        )
        
    # 2. 입체 도형의 꼭짓점 8개의 사영 위치를 Accent Green(#06d6a0) 원형 점(마커)으로 렌더링합니다.
    # 안티 브루탈리즘 테마에 맞추어 2px 굵기의 명확한 검정 테두리를 둘러 입체적 질감 포인트를 부여합니다.
    ax.plot(projected_x, projected_y, 'o', color='#06d6a0', markersize=10, markeredgecolor='black', markeredgewidth=2, 
            label='입체 정점 (Projected Vertices)')
    
    # 2D 사영 스크린의 x축, y축 눈금선 중심 기준을 2.5px 굵은 검정 실선으로 그어 투영의 평면 대칭 구조를 시각화합니다.
    ax.axhline(0, color='black', linewidth=2.5)
    ax.axvline(0, color='black', linewidth=2.5)
    
    # 연한 회색 그리드선을 보조로 얹습니다.
    ax.grid(True, color='#e0e0e0', linestyle='--', linewidth=1)
    
    # 뷰포트 비율을 1:1 대칭 고정하고 범위를 강제하여 슬라이더 회전 시 박스가 쪼그라들거나 비율이 찌그러지는 왜곡을 방지합니다.
    ax.set_aspect('equal')
    ax.set_xlim(-6, 6)
    ax.set_ylim(-6, 6)
    
    # 3. 투영 조건 요약 정보(범례) 설정: Vibrant Yellow (#ffd166) 배경에 2px 검정 테두리를 둘러 프리미엄 디자인 룩 매치
    legend = ax.legend(loc='upper right', fontsize=11, frameon=True)
    frame = legend.get_frame()
    frame.set_facecolor('#ffd166')
    frame.set_edgecolor('black')
    frame.set_linewidth(2)
    
    # 차트 주변 굵은 테두리(Spines) 설정 (3px 검정)
    for spine in ax.spines.values():
        spine.set_color('black')
        spine.set_linewidth(3)
        
    # 차트 설명 텍스트 렌더링
    ax.set_title(f'3D 원근 투영: Pitch={pitch_deg}°, Yaw={yaw_deg}°, Roll={roll_deg}° (카메라 거리 d={d_camera:.1f})', 
                 fontsize=13, fontweight='bold', pad=15)
    ax.set_xlabel('투영 캔버스 X-screen (px)', fontsize=11)
    ax.set_ylabel('투영 캔버스 Y-screen (px)', fontsize=11)
    
    plt.show()
```

### 🎬 3단계: 3D 원근 투영 시뮬레이터 구동 (Plotting & Execution)
카메라의 회전 각도 세 방향(Pitch, Yaw, Roll) 슬라이더와 원근 거리 인자 $d$의 슬라이더를 마우스로 조작해 보면서,
3차원 정육면체가 카메라의 위치와 각도 변화에 따라 2차원 평면 위에 맺히는 그림자가 어떻게 상호 변환을 이루는지 기하학적으로 고찰해 봅니다.

```python
# interact 함수를 사용하여 3D 원근 투영 시뮬레이터 함수에 조절 슬라이더 4개를 연동시킵니다.
# 회전 각도들은 -180도에서 180도까지 5도 단위로 자유롭게 회전 조절할 수 있습니다.
# 카메라 거리 d_camera는 3.0에서 15.0까지 조절하여 원근 왜곡 효과(Perspective distortion)를 관찰할 수 있게 구성합니다.
interact(plot_perspective_projection, 
         pitch_deg=widgets.FloatSlider(value=30.0, min=-180.0, max=180.0, step=5.0, description='회전 Pitch (x)'),
         yaw_deg=widgets.FloatSlider(value=45.0, min=-180.0, max=180.0, step=5.0, description='회전 Yaw (y)'),
         roll_deg=widgets.FloatSlider(value=0.0, min=-180.0, max=180.0, step=5.0, description='회전 Roll (z)'),
         d_camera=widgets.FloatSlider(value=6.0, min=3.0, max=15.0, step=0.5, description='원근 거리 (d)'))
```
